```python
import pandas as pd
```

```python
n225=pd.read_csv('N225.csv').dropna()
```

```python
n225.describe()
```

```python
def f(row):
    r='Up' if row['Close']>row['Open'] else 'Down'
    return r
n225['Direction']=n225.apply(f,axis=1)
n225['Range']=n225['High']-n225['Low']
n225['Open Down']=n225['Open'] > n225['Close'].shift()
```

```python
n225.head()
```

```python
n225.to_csv("N225_clean.csv")
```
